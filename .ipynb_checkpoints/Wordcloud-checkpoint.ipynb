```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = """Short text to present wordcloud commas and dots are removed and word wordcloud is 
repeated couple of times to show nice plot of wordcloud don't mind wordcloud wordcloud wordcloud wordcloud wordcloud wordcloud 
wordcloud text text text"""
```

```python
wordcloud = WordCloud(max_font_size=40, collocations=False).generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
